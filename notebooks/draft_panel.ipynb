```python
import numpy as np
from statsmodels.datasets import grunfeld
data = grunfeld.load_pandas().data
data.year = data.year.astype(np.int64)
data = data.set_index(['firm', 'year'])

from linearmodels import PanelOLS, RandomEffects
from linearmodels.panel import PooledOLS
from linearmodels.panel.model import PanelOLS
from linearmodels.panel.model import RandomEffects
from linearmodels.panel.model import PooledOLS
from linearmodels.panel import PanelOLS, RandomEffects
from linearmodels.panel.model import PooledOLS
from linearmodels.panel.model import PanelOLS
from linearmodels.panel.model import RandomEffects
from linearmodels.panel.model import PooledOLS
from linearmodels.panel.model import PanelOLS
import pandas as pd
import numpy as np
from statsmodels.iolib.summary2 import summary_col
```

```python
# Pooled OLS
exog_vars = ['value', 'capital']
exog = sm.add_constant(data[exog_vars])
pooled_ols = PooledOLS(data.invest, exog)
pooled_ols_res = pooled_ols.fit()
print(pooled_ols_res)
```

```python
# Fixed Effects
mod_fixed = PanelOLS(data.invest, data[['value', 'capital']], entity_effects=True)
res_fixed = mod_fixed.fit(cov_type='clustered', cluster_entity=True)
print(res_fixed)
```

```python
# Random Effects
mod_random = RandomEffects(data.invest, data[['value', 'capital']])
res_random = mod_random.fit()
print(res_random)
```

```python
# Dynamic Panel Data Models - Arellano-Bond
from linearmodels.panel import PanelOLS, compare

# Add lagged dependent variable
data['invest_lag1'] = data.groupby('firm')['invest'].shift(1)
data.dropna(inplace=True)

# Arellano-Bond Estimator
mod_ab = PanelOLS(data.invest, data[['invest_lag1', 'value', 'capital']], entity_effects=True)
res_ab = mod_ab.fit(cov_type='clustered', cluster_entity=True)
print(res_ab)
```

```python
# Comparison of models
results = compare({'Pooled OLS': pooled_ols_res, 'Fixed Effects': res_fixed, 'Random Effects': res_random, 'Arellano-Bond': res_ab})
print(results)
```

```python
import numpy as np
import pandas as pd
from statsmodels.datasets import grunfeld
from linearmodels import PanelOLS, RandomEffects
from linearmodels.panel import compare
from linearmodels.panel.model import RandomEffects
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt

# Load the data
data = grunfeld.load_pandas().data
data.year = data.year.astype(np.int64)
data = data.set_index(['firm', 'year'])

# Train-test split
train_data = data[data.index.get_level_values('year') < 1986]
test_data = data[data.index.get_level_values('year') >= 1986]

# Define function to calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Define function to fit model and calculate error metrics
def fit_model_and_evaluate(model, train_data, test_data):
    results = model.fit(cov_type='clustered', cluster_entity=True)
    predictions = results.predict(exog=test_data[['value', 'capital']])
    
    mae = mean_absolute_error(test_data['invest'], predictions)
    mse = mean_squared_error(test_data['invest'], predictions)
    rmse = sqrt(mse)
    mape = mean_absolute_percentage_error(test_data['invest'], predictions)
    
    return results, mae, mse, rmse, mape

# Pooled OLS model
pooled_ols = PanelOLS(train_data['invest'], train_data[['value', 'capital']], entity_effects=False)
pooled_ols_results, pooled_ols_mae, pooled_ols_mse, pooled_ols_rmse, pooled_ols_mape = fit_model_and_evaluate(pooled_ols, train_data, test_data)

# Fixed Effects model
fixed_effects = PanelOLS(train_data['invest'], train_data[['value', 'capital']], entity_effects=True)
fixed_effects_results, fixed_effects_mae, fixed_effects_mse, fixed_effects_rmse, fixed_effects_mape = fit_model_and_evaluate(fixed_effects, train_data, test_data)

# Random Effects model
random_effects = RandomEffects(train_data['invest'], train_data[['value', 'capital']])
random_effects_results, random_effects_mae, random_effects_mse, random_effects_rmse, random_effects_mape = fit_model_and_evaluate(random_effects, train_data, test_data)

# Dynamic Panel Data Model - Arellano-Bond
train_data['invest_lag1'] = train_data.groupby('firm')['invest'].shift(1)
test_data['invest_lag1'] = test_data.groupby('firm')['invest'].shift(1)

dynamic_panel = PanelOLS(train_data['invest'], train_data[['value', 'capital', 'invest_lag1']].dropna(), entity_effects=True)
dynamic_panel_results, dynamic_panel_mae, dynamic_panel_mse, dynamic_panel_rmse, dynamic_panel_mape = fit_model_and_evaluate(dynamic_panel, train_data.dropna(), test_data.dropna())

# Summarize Results
summary_table = pd.DataFrame({
    'Model': ['Pooled OLS', 'Fixed Effects', 'Random Effects', 'Arellano-Bond'],
    'MAE': [pooled_ols_mae, fixed_effects_mae, random_effects_mae, dynamic_panel_mae],
    'MSE': [pooled_ols_mse, fixed_effects_mse, random_effects_mse, dynamic_panel_mse],
    'RMSE': [pooled_ols_rmse, fixed_effects_rmse, random_effects_rmse, dynamic_panel_rmse],
    'MAPE': [pooled_ols_mape, fixed_effects_mape, random_effects_mape, dynamic_panel_mape]
})

print(summary_table)
print(compare({'Pooled OLS': pooled_ols_results, 'Fixed Effects': fixed_effects_results, 'Random Effects': random_effects_results, 'Arellano-Bond': dynamic_panel_results}))

# Note: Due to the introduction of NaNs for lags, dropping NaNs is necessary to ensure the comparison is accurate.
```
