```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_excel(r'..\data\raw\Sri Lankan Ruppee to USD exchange rate.xls')
df1.head(15)
```

```python
df2=pd.read_excel(r'..\data\raw\Singapore Dollar to USD exchange rate.xls')
df2.head(15)
```

```python
df3=pd.read_excel(r'..\data\raw\Indian Ruppee to USD exchange rate.xls')
df3.head(15)
```

```python
df4=pd.read_excel(r'..\data\raw\Japan Yen to USD exchange rate.xls')
df4.head(15)
```

```python
# remove first 10 rows
df1 = df1.iloc[10:]
df2 = df2.iloc[10:]
df3 = df3.iloc[10:]
df4 = df4.iloc[10:]
```

```python
# rename columns to Date and Currency(USD)
df1.columns = ['Date', 'Currency(USD)']
df2.columns = ['Date', 'Currency(USD)']
df3.columns = ['Date', 'Currency(USD)']
df4.columns = ['Date', 'Currency(USD)']
```

```python
# create new column for Country
df1['Country'] = 'Sri Lanka'
df2['Country'] = 'Singapore'
df3['Country'] = 'India'
df4['Country'] = 'Japan'
```

```python
df3
```

```python
# concatenate all dataframes
df = pd.concat([df1, df2, df3, df4])
df
```

```python
df=df[['Country','Date', 'Currency(USD)']]
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
# if currency have 0 value, replace with NaN
df['Currency(USD)'] = df['Currency(USD)'].replace(0, np.nan)
df.isnull().sum()
```

```python
df[df['Currency(USD)'].isnull()]
```

```python
df['Country'].value_counts()
```

```python
# reshape data daily
df = df.pivot(index='Date', columns='Country', values='Currency(USD)')
df
```

```python
df.isnull().sum()
```

```python
# plot missing values of each column separately
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

```python
#  remove date before 2019-06-03
df = df.loc['2019-06-03':]
df
```

```python
# plot missing values of each column separately
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

```python
df.isnull().sum()
```

```python
# unpivot dataframe
df = df.melt(ignore_index=False, value_name='Currency(USD)').reset_index()
df
```

```python
# fill missing value with average of previous and next value
df['Currency(USD)'] = df.groupby('Country')['Currency(USD)'].transform(lambda x: x.fillna(x.interpolate()))
df.isnull().sum()
```

```python
# plot how Currency(USD) varies over time for each country in separate subplots using loop
countries = df['Country'].unique()
fig, axes = plt.subplots(nrows=len(countries), figsize=(10, 20))
for i, country in enumerate(countries):
    df[df['Country'] == country].plot(x='Date', y='Currency(USD)', ax=axes[i], title=country)
plt.tight_layout()
plt.show()
```

```python
# in Currency(USD) column, put the value of 1/Currency(USD)
df['Currency(USD)'] = 1/df['Currency(USD)']
df
```

```python
# plot how Currency(USD) varies over time for each country in separate subplots using loop
countries = df['Country'].unique()
fig, axes = plt.subplots(nrows=len(countries), figsize=(10, 20))
for i, country in enumerate(countries):
    df[df['Country'] == country].plot(x='Date', y='Currency(USD)', ax=axes[i], title=country)
plt.tight_layout()
plt.show()
```

```python
df=df[['Country','Date', 'Currency(USD)']]
```
