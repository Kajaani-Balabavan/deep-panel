```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vaccination = pd.read_csv(r'..\data\Base_Paper\Deep Neural Network Estimation in Panel Data Models\vaccinations.csv')
vaccination.head()
```

```python
positive_rate = pd.read_csv(r'..\data\Base_Paper\Deep Neural Network Estimation in Panel Data Models\positive-rate-daily-smoothed.csv')
positive_rate.head()
```

```python
# drop Code column
positive_rate = positive_rate.drop(columns=['Code'])
positive_rate.head()
```

```python
who= pd.read_csv(r'..\data\Base_Paper\Deep Neural Network Estimation in Panel Data Models\WHO-COVID-19-global-data.csv')
who.head()
```

```python
who.columns
```

```python
# drop 'Country_code'
who = who.drop(columns=['Country_code', 'WHO_region'])
who.head()
```

```python
test_per_thousand = pd.read_csv(r'..\data\Base_Paper\Deep Neural Network Estimation in Panel Data Models\full-list-cumulative-total-tests-per-thousand.csv')
test_per_thousand.head()
```

```python
OxCGRT = pd.read_csv(r'..\data\Base_Paper\Deep Neural Network Estimation in Panel Data Models\OxCGRT_compact_subnational_v1.csv')
OxCGRT.head()
```

```python
# in OxCGRT, the column date is in the format YYYYMMDD. We need to convert it to a datetime format
OxCGRT['Date'] = pd.to_datetime(OxCGRT['Date'], format='%Y%m%d')
OxCGRT.head()
```

```python
# in every dataset, filter the data for the country of interest Canada,France, Germany, Italy, Japan, United Kingdom, United States

vaccination = vaccination[vaccination['location'].isin(['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'])]
positive_rate = positive_rate[positive_rate['Entity'].isin(['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'])]
who = who[who['Country'].isin(['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'])]
test_per_thousand = test_per_thousand[test_per_thousand['Entity'].isin(['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'])]
OxCGRT = OxCGRT[OxCGRT['CountryName'].isin(['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States'])]
```

```python
# for all datasets filter January 1, 2020, to December 24, 2022
vaccination = vaccination[(vaccination['date'] >= '2020-01-01') & (vaccination['date'] <= '2022-12-24')]
positive_rate = positive_rate[(positive_rate['Day'] >= '2020-01-01') & (positive_rate['Day'] <= '2022-12-24')]
who = who[(who['Date_reported'] >= '2020-01-01') & (who['Date_reported'] <= '2022-12-24')]
test_per_thousand = test_per_thousand[(test_per_thousand['Day'] >= '2020-01-01') & (test_per_thousand['Day'] <= '2022-12-24')]
OxCGRT = OxCGRT[(OxCGRT['Date'] >= '2020-01-01') & (OxCGRT['Date'] <= '2022-12-24')]
```

```python
columns = [
    'Country',
    'Date',
    'StringencyIndex_Average', 
    'C1M_School.closing', 
    'C2M_Workplace.closing', 
    'C3M_Cancel.public.events', 
    'C4M_Restrictions.on.gatherings',
    'C5M_Close.public.transport',
    'C6M_Stay.at.home.requirements',
    'C7M_Restrictions.on.internal.movement',
    'C8EV_International.travel.controls',
    'H1_Public.information.campaigns', 
]
```

```python
# filter the columns
OxCGRT = OxCGRT[columns]
OxCGRT.head()
```

```python
OxCGRT[(OxCGRT['Country'] == 'Canada') & (OxCGRT['Date'] == '2020-01-01')]
```

```python
# for each country,date pair, we have multiple entries. We need to aggregate the data. use and logic. if all values are 1 then 1 else 0
OxCGRT = OxCGRT.groupby(['Country', 'Date']).agg(lambda x: 1 if (x == 1).all() else 0).reset_index()
OxCGRT.head()
```

```python
OxCGRT[(OxCGRT['Country'] == 'Canada') & (OxCGRT['Date'] == '2020-01-01')]
```

```python
# merge all datasets on 'Country' and 'Date'
data = vaccination.merge(positive_rate, on=['Country', 'Date'], how='left')
data = data.merge(who, on=['Country', 'Date'], how='left')
data = data.merge(test_per_thousand, on=['Country', 'Date'], how='left')
data = data.merge(OxCGRT, on=['Country', 'Date'], how='left')
data.head()
```
